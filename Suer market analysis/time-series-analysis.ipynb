# Using the correct format

Time series data can come in with so many different formats. But not all of those formats are friendly to python’s pandas' library. The most convenient format is the timestamp format for Pandas. But most of the time time-series data come in string formats.



dates = ['2020-11-25 2:30:00 PM', 'Jan 5, 2020 18:45:00', '01/11/2020', '2020.01.11', '2020/01/11', '20201105']

The ‘dates’ variable above showing five different formats of date-time settings and all are correct. But we need this specific format to work conveniently. So, convert those dates to the right format.

```python
df["Time"]= pd.to_datetime(df["Time"])
df.head()
```
